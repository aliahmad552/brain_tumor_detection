```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten,Conv2D,MaxPooling2D,Dropout,BatchNormalization
```

```python
data_gen = ImageDataGenerator(
    rescale = (1./255)
)
```

```python
train_ds = data_gen.flow_from_directory(
    base_dir + '/Training',
    target_size = (224,224),
    batch_size = 32,
    class_mode = 'categorical'
)
```

```python
test_ds = data_gen.flow_from_directory(
    base_dir+'/Testing',
    batch_size = 32,
    target_size = (224,224),
    class_mode = 'categorical'

)
```

```python
images,labels = next(test_ds)
```

```python
class_names = list(train_ds.class_indices.keys())

plt.figure(figsize=(12, 8))
for i in range(9):  # show first 9 images
    plt.subplot(3, 3, i + 1)
    plt.imshow(images[i])
    plt.title(class_names[labels[i].argmax()])  # convert one-hot to class name
    plt.axis('off')

plt.tight_layout()
plt.show()
```

# CNN

```python
model = Sequential()
model.add(Conv2D(32,(3,3),padding = 'valid',activation = 'relu',input_shape = (224,224,3)))
model.add(Dropout(0.3))
model.add(Conv2D(64,(3,3),padding = 'valid',activation = 'relu'))
model.add(Dropout(0.3))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(64,(3,3),padding = 'valid',activation = 'relu'))
model.add(Dropout(0.3))
model.add(Conv2D(128,(3,3),padding = 'valid',activation = 'relu'))
model.add(Dropout(0.3))
model.add(MaxPooling2D(2,2))

model.add(Flatten())
model.add(Dense(256,activation = 'relu'))
model.add(Dropout(0.3))
model.add(Dense(len(class_names),activation = 'softmax'))

model.summary()
```

```python
model.compile(optimizer = 'adam',loss = 'categorical_crossentropy',metrics = ['accuracy'])
```

```python
history = model.fit(train_ds,epochs = 10,validation_data = test_ds)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.show()
```
